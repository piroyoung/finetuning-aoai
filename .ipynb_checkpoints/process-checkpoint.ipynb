```python
import pandas as pd
import json
```

```python
pd.read_csv("kana.csv")
```

```python
pd.read_csv("kanji.csv")
```

```python
kana = pd.read_csv("kana.csv")
```

```python
kanji = pd.read_csv("kanji.csv")
```

```python
df = pd.concat([kana, kanji]).assign(
    prompt=lambda x: x.kami1 + " " + x.kami2 + " " + x.kami3,
    completion=lambda x: x.shimo1 + " " + x.shimo2
).pipe(
    lambda df: df[["prompt", "completion"]]
).pipe(lambda df: df.reset_index()[["prompt", "completion"]])
```

```python
with open("corpus.jsonl", mode="w", encoding="utf_8") as f:
    for index, row in df.iterrows():
        f.write(
            json.dumps(
            {
                "prompt": row.prompt,
                "completion": row.completion
            }, ensure_ascii=False
            )
        )
        f.write("\n")
    
```
